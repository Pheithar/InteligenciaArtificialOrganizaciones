```python
import pandas as pd
from langdetect import detect
```

```python
data = pd.read_csv("./datos/book_data.csv")
```

```python
display(data)
```

```python
def removeStopWords(text):
    
    stopwords = []
    
    with open('datos/stopwords.txt') as f:
        for i in f.readlines():
            stopwords.append(i[:-1])
    
    text = text.replace('.', ' ')
    text = text.replace(',', ' ')
    text = text.replace('-', ' ')
    text = text.replace('_', ' ')
    text = text.replace('–', ' ')
    text = text.replace('•', ' ')
    text = text.replace('(', ' ')
    text = text.replace(')', ' ')
    text = text.replace('[', ' ')
    text = text.replace(']', ' ')
    text = text.replace(';', ' ')
    text = text.replace('{', ' ')
    text = text.replace('}', ' ')
    text = text.replace('?', ' ')
    text = text.replace('!', ' ')
    text = text.replace('"', ' ')
    text = text.replace('”', ' ')
    text = text.replace('\n', ' ')
    text = text.replace('\t', ' ')
    text = text.replace('\r', ' ')
    text = text.replace('<br>', ' ')
    
    text = text.lower()
    
    while '  ' in text:
        text = text.replace('  ', ' ')
    
    splitted_text = text.split()
    
    new_splitted_text = [word for word in splitted_text if not word in stopwords]
    
    new_text = ' '.join(new_splitted_text)

    
    
    return new_text
```

```python
newData = []
test = data[["book_desc", "genres"]]

for desc, genres in data[["book_desc", "genres"]].values:
    try:
        if detect(desc) == 'en' and type(genres) == type(''):
            desc = removeStopWords(desc)
            newData.append([desc, genres])
            
    except:
        pass
```

```python
newData = pd.DataFrame(newData, columns=["Description", "Genres"])
```
